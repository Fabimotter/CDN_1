```python
def primo (numero):
  eprimo=False
  resto_zero=0

  for i in range (1,numero):
    if (numero%1)==0:
      resto_zero += 1

  if resto_zero==1:
    eprimo= True

    return eprimo
n=int(input (" digite um valor inteiro e positivo e maior que zer: "))

if primo (n):
  print (f"O numero {n} é primo")
else:
  print (f"O numero {n} não é primo")
```
