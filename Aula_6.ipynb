```python
#FATORIAL
numero=int(input("Digite um número: "))
fatorial=1
if numero<0:
  print("Fatorial não existe")
elif numero==0:
  print("Fatorial de zero é igual 1")
else:
  for i in range(1,numero+1):
    fatorial= fatorial*i
  print(" O fatorial de ", numero, "é igual", fatorial)
#
```
