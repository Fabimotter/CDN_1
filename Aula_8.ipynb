```python
import random

jogar_novamente = "s"

while jogar_novamente == "s":

    print("Bem-vindo ao jogo de adivinhação!")

    print("Escolha o nível de dificuldade:")
    print("1 - Fácil: 10 tentativas, número de 1 a 50")
    print("2 - Difícil: 5 tentativas, número de 1 a 100")

    nivel = int(input("Digite 1 para fácil ou 2 para difícil: "))

    if nivel == 1:
        limite = 10
        numero_maximo = 50
    else:
        limite = 5
        numero_maximo = 100

    numero_secreto = random.randint(1, numero_maximo)
    tentativas = 0
    acertou = False

    while tentativas < limite:
        print("Tentativa", tentativas + 1, "de", limite)

        palpite = int(input("Tente adivinhar o número: "))
        tentativas = tentativas + 1

        diferenca = abs(palpite - numero_secreto)

        if palpite == numero_secreto:
            print("Acertou!")
            print("Você usou", tentativas, "tentativa(s).")
            acertou = True
            break

        elif palpite < numero_secreto:
            print("O número secreto é maior!")
        else:
            print("O número secreto é menor!")

        if diferenca <= 2:
            print("Você está quente!")
        elif diferenca > 20:
            print("Você está frio...")
        else:
            print("Você está morno.")

    if acertou == False:
        print("Você perdeu!")
        print("O número secreto era", numero_secreto)

    jogar_novamente = input("Deseja jogar de novo? Digite s para sim ou n para não: ")


```
